```python
import pickle
from pathlib import Path
import numpy as np

dir = Path("CIFAR-10-batches-py/")

def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

output = []
for batch in dir.glob("data_batch_*"):
    photos = unpickle(batch)
    output += list(np.asarray(photos[b"data"]))
```

```python
from PIL import Image

input = []
for img in output:
    tmp = np.swapaxes(img.reshape(3,32,32), 0, 2)
    input.append(np.asarray(Image.fromarray(tmp, "RGB").convert('L')).reshape(1024))
```

```python
import pandas as pd
df = pd.DataFrame({"input":input, "output":output})
df
```

```python
import matplotlib.pyplot as plt
start = 25000
for i in range(50):
    ax = plt.subplot(5, 10, i + 1)
    if i % 2 == 0:
        df.loc[i]["input"]
        plt.imshow(df.loc[start + i]["input"].reshape(32,32), cmap = "gray")
    else:
        img = df.loc[start + i-1]["output"].reshape(3,32,32)
        img = np.swapaxes(img, 0, 2)
        plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
```
